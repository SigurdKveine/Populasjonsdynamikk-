a*b(t) er fødselnummeret til arten, med a som en reproduksjonsrate. b*b(t)r(t) er hvor mange som blir drept av rovdyret vi har valgt, hvor b er en konstant som kan avgjøres på flere måtes. c*b(t)r(t) er en fødselsledd for rovdyret, som avhenger av antall byttedyr og en reproduksjonsrate c. c*r(t) er hvor mange rovdyr som dør, med en konstant "dødstall" c. 

2) 

```python
from pylab import *
N = 100000
tid = 2
dt = tid/(N-1)
R = zeros(N)
B = zeros(N)
Bder = zeros(N)
Rder = zeros(N)
R[0] = 100
B[0] = 100
t = zeros(N)

a = 0.25
b = 0.002
c = 0.1
d = 0.001

for i in range (N-1):
    Bder[i] = a*B[i] - b*B[i]*R[i]
    Rder[i] = c*R[i]*B[i] - d*R[i]
    B[i+1] = B[i] + Bder[i]*dt
    R[i+1] = R[i] + Rder[i]*dt
    t[i+1] = t[i]+dt
    
x = t
y_1 = R
y_2 = B
plot(t,R)
plot(t,B)
ylim(0,500)
grid("on")
show()
```
